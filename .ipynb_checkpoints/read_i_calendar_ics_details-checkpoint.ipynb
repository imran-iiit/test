```python
# ### Scrub data to get the instructor name!!!
# import requests
# from bs4 import BeautifulSoup 

# URL = 'https://learning.oreilly.com/live-training/courses/introduction-to-statistics-for-data-analysis-with-python/0636920308287/'
# URL = 'https://learning.oreilly.com/live-training/courses/introduction-to-algorithms-and-data-structures/0636920321644/'

def get_presenter(URL):
    r = requests.get(URL) 
    # print(r.content) 


    soup = BeautifulSoup(r.content, 'html5lib') 
    # print(soup.prettify()) 

    # print(soup.find('meta', attrs = {'property':'og:title'}) )
    # print(soup.script.context)
    # table

    name = soup.find('span', attrs = {'class':'name', 'itemprop':'name'})
    return name.string if name else ''
```

```python
def get_ics_data(file_name):
    with open(file_name,'rb') as g:
        gcal = Calendar.from_ical(g.read())
        for component in gcal.walk():
            if component.name == "VEVENT":
                summary = component.get('summary')
                dtstart = component.get('DTSTART').dt.astimezone(get_localzone()) #.strftime('%Y-%m-%d %H:%M:%S')
                dtend   = component.get('dtend').dt.astimezone(get_localzone()) #.strftime('%Y-%m-%d %H:%M:%S')
        #         print(component.get('dtstamp').dt.astimezone(get_localzone()))
                desc = component.decoded('DESCRIPTION')
#                 print(desc)
        #         print(dir(component))
                urls = re.findall('https:.*[0-9]+', str(desc))
                url = urls[0] if urls else ''
#                 print(urls)
#                 print(url)
                presenter = get_presenter(url)
                ind = 'Asia/Kolkata'
                start = component.get('DTSTART').dt
#                 print(start)
                local_start = start.astimezone(get_localzone()) #.strftime('%Y-%m-%d %H:%M:%S')
                
                return [file_name, summary, dtstart, dtend, url, '; '.join(urls), presenter, local_start, desc]
                
#     g.close()
```
