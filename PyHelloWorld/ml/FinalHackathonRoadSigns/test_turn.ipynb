```python
import torch
turn_model = torch.load('i_turn.pth.tar')
```

```python
import numpy as np
from PIL import Image
from torchvision import utils
from torchvision import transforms
from torch.autograd import Variable

# from torchvision.utils import recreate_image

def normalize(image):
    normalize = transforms.Normalize(
                                        mean=[0.485, 0.456, 0.406],
                                        std=[0.229, 0.224, 0.225]
                                    )
    preprocess = transforms.Compose([
                                        transforms.Resize((224,224)),
                                        transforms.ToTensor(),
                                        normalize
                                    ])
    image = Variable(preprocess(image).unsqueeze(0))
    return image

def predict_turn(raw_img):
    CLASSES = ['left', 'misc', 'right', 'signal', 'stop', 'u_turn']
    img = normalize(raw_img)
    out_v = turn_model(img)
    
    return CLASSES[np.argmax(out_v.detach().numpy())]
```

```python
cv2_img = Image.open('camera_img.jpeg')
h = predict_turn(cv2_img)

print(h)
```
